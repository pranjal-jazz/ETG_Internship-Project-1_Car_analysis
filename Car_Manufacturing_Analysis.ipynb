Brands & No. of brands in the datsaet

```python
print("No. of brands =", len(df['brand'].unique()))
data = []

for brands in df['brand'].unique():
  data.append([brands, len(df[df['brand'] == brands])])
data =pd.DataFrame(data, columns= ['brands', 'freq'])
data.sort_values(by= 'freq', ascending = False)
```

```python
data.plot(x= 'brands' , y= 'freq', figsize= (5,5))
```

Models & No. of models

```python
print("No. of models =", len(df['model'].unique()))
data = []

for models in df['model'].unique():
  data.append([models, len(df[df['model'] == models])])
data =pd.DataFrame(data, columns= ['models', 'freq'])
data.sort_values(by= 'freq', ascending = False)

```

```python
data.plot(x= 'models' , y= 'freq', kind='bar', figsize= (40,20))
```

```python
data = []

for years in df['year'].unique():
  data.append([years, len(df[df['year'] == years])])
data =pd.DataFrame(data, columns= ['years', 'freq'])
data.sort_values(by= 'freq', ascending = True)
```

Graph of brand having maximum average price

```python
df.groupby('brand').mean()['price'].sort_values(ascending = False).plot.bar(figsize =(10,5))
```

Some more data visualization

```python
grp =df.groupby('brand')
x = grp['price'].agg(np.mean)
z = grp['mileage'].agg(np.mean)
print(x)
print(z)
```

Now Let's have some comparison 

```python
plt.figure(figsize = (15,6))
plt.plot(x, "r--", color='green')
plt.xticks(rotation=270)
plt.title('brand vs price ')
plt.xlabel('brand')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize = (15,6))
plt.plot(z, "ro", color='green')
plt.xticks(rotation=90)
plt.title('brand vs mileage ')
plt.xlabel('brand')
plt.ylabel('mileage')
plt.show()
```

```python
grp =df.groupby('year')
x = grp['price'].agg(np.mean)
z = grp['mileage'].agg(np.mean)
```

```python
plt.figure(figsize = (15,6))
plt.plot(x, "r", color='yellow')
plt.xticks(rotation=90)
plt.title('year & price ')
plt.xlabel('year')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize = (15,6))
plt.plot(z, "ro--", color='red')
plt.xticks(rotation=90)
plt.title('year & mileage ')
plt.xlabel('year')
plt.ylabel('mileage')
plt.show()
```
